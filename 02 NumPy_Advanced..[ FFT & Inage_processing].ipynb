#####  **Bitwise Operations on NumPy Arrays**

```python
x = np.array([0, 1, 0, 1], np.uint8)
y = np.array([0, 0, 1, 1], np.uint8)

print(np.bitwise_and(x, y))
print(np.bitwise_or(x, y))
print(np.bitwise_xor(x, y))
print(np.bitwise_not(x))
```

##### **Write a Function That Takes Two NUMPY Matrices and Return The ADDITION , SUBTRACTION ,TRANsPOSE ,PRODUCT ,DOT PRODUCT**

```python
def matrix_operation(a,b):
    print('Matrices addition = \n ',np.add(a,b))
    print('Matrices Subtract =  \n',np.subtract(a,b))
    print('Matrices multiply =  \n',np.multiply(a,b))
    print('Matrices Transpose =  \n',a.T)
    print('Matrices Transpose =  \n',b.T)
    print('Matrices DOT_MULTIPLi = \n ',np.dot(a,b))
    print('Matrices squre_rout = \n',np.sqrt(a))
```

```python
x = np.array([[3,9],
             [8,5]])

y = np.array([[2,3],
             [1,7]])

matrix_operation(x,y)
```

#### **Fast Fourier Transform**

```python
import numpy as np
import matplotlib.pyplot as plt
```

##### **1-D FFT**

```python
t = np.arange(128)
print(t)
```

```python
sp = np.fft.fft(np.sin(t))
freq = np.fft.fftfreq(t.shape[-1])
print(sp,freq)
```

```python
plt.subplot(2,1,1)
plt.title('Real Part')
plt.plot(freq,sp.real)

plt.subplots_adjust(hspace = 0.5)

plt.subplot(2,1,2)
plt.title('Imaginary Part')
plt.plot(freq,sp.imag)

plt.show()
```

##### **Inverse 1-D DFT**

```python
t = np.arange(300)
n = np.zeros((300, ),dtype = complex)
print(t,n)
```

```python
n[230:270] = np.exp(1j*np.random.uniform(0,2*np.pi, (40, )))
s = np.fft.ifft(n)
```

```python
plt.subplot(2,1,1)
plt.title('Real Part')
plt.plot(t,s.real)

plt.subplots_adjust(hspace = 0.5)

plt.subplot(2,1,2)
plt.title('Imaginary Part')
plt.plot(t,s.imag)

plt.show()
```

```python
plt.title('IFFT')
plt.plot(t,s.real)
plt.plot(t,s.imag)

plt.show()
```
